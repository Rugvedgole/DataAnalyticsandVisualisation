```python
import pandas as pd
from time import strftime
from datetime import datetime
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("showwcase_sessions.csv")
```

```python
df = df.dropna()
df
```

# Customer Analysis & Date Analysis

```python
def appendCustomerCount(df):
    customer_engagement = df['customer_id'].value_counts().to_frame().reset_index().rename(columns={"index": "customer_id", "customer_id": "customer_count"})
    return customer_engagement
```

```python
def appendCustomerActivity(df):
    df1= df[['customer_id','session_projects_added','session_likes_given','session_comments_given']]
    df1 = df1.groupby('customer_id').sum().reset_index()
    df1['activity_score'] = df1['session_projects_added']+df1['session_likes_given']+df1['session_comments_given']
    df1=df1[['customer_id','activity_score']]    
    return df1
```

```python
def appendCustomerProjLikeComment(df):
    df1= df[['customer_id','session_projects_added','session_likes_given','session_comments_given','inactive_duration','bugs_in_session','session_duration']]
    df1 = df1.groupby('customer_id').sum().reset_index()   
    return df1
```

```python
customer_engagement = appendCustomerCount(df)
customer_activity = appendCustomerActivity(df)
customer_proj_like_comment= appendCustomerProjLikeComment(df)
customer_engagement = pd.merge(customer_engagement, customer_activity, how='inner', on=['customer_id', 'customer_id'])
customer_engagement = pd.merge(customer_engagement, customer_proj_like_comment, how='inner', on=['customer_id', 'customer_id'])
customer_engagement['net_active_duration']=customer_engagement['session_duration']-customer_engagement['inactive_duration']
#customer_engagement
customer_engagement.sort_values(by = 'activity_score', ascending = False)
```

# Retention Rate

```python
#Assigning Start date as the first date the customer visited the website, to find the Retention rate
```

```python
df3= df[['customer_id','login_date']]
df3 = df3.rename(columns = {'login_date':'start_date'})
df4= df[['customer_id','login_date']]
df4 = df4.rename(columns = {'login_date':'session_date'})

df3.start_date = pd.to_datetime(df3.start_date, format='%m/%d/%y')
df4.session_date = pd.to_datetime(df4.session_date, format='%m/%d/%y')


df3 = df3.groupby('customer_id').min()['start_date'].reset_index()
new = pd.merge(df3, df4, how='inner', on=['customer_id', 'customer_id'])



new['seniority'] = (new['session_date'] - new['start_date']).dt.days + 1
```

```python
new
```

```python
group = new.groupby(['start_date','seniority'])

cohort_data = group['customer_id'].size()

cohort_data = cohort_data.reset_index()

cohort_data
```

```python

cohort_counts = cohort_data.pivot(index='start_date', columns='seniority', values='customer_id')

# We need the base for our retention rate calculation
# Here we just take the first column of cohort_counts
base = cohort_counts[1]

# And now we apply the divide method to our cohort_counts, dividing it by the base
retention = cohort_counts.divide(base, axis=0).round(3)
```

```python
# User Count heatmap
plt.figure(figsize=(18,14))
plt.title('Users Active')
ax = sns.heatmap(data=cohort_counts, annot=True, vmin=0.0,cmap='Reds')
ax.set_yticklabels(cohort_counts.index)
fig=ax.get_figure()
fig.savefig("Retention Counts.png")

plt.show()
```

```python
# Retention Rate heatmap
plt.figure(figsize=(18,14))
plt.title('Retention Table')
ax = sns.heatmap(data=retention, annot=True, fmt='.0%', vmin=0.0, vmax=1,cmap='Reds')
ax.set_yticklabels(retention.index)
fig=ax.get_figure()
fig.savefig("Retention_Rate.png")

plt.show()
```
